```python
import pandas as pd
```

```python
data = pd.read_csv('.././csv/OnlineRetail.csv',encoding='latin1')
df = pd.DataFrame(data=data)
df.head(5)
```

```python
df.info()
```

```python
#what is the unique count of the invoice numbers:
len(pd.unique(df['InvoiceNo']))
#25900
```

```python
#How many people bought “PARTY BUNTING”.:*
df[df['Description']=='PARTY BUNTING']['Description'].count()
#1727
```

```python
#How many people bought “WHITE HANGING HEART T-LIGHT HOLDER”.
df[df['Description']=='WHITE HANGING HEART T-LIGHT HOLDER']['Description'].count()
#2369
```

```python
#how many unique people are from “UNITED KINGDOM”
df[df['Country']=='United Kingdom']['CustomerID']
df[['CustomerID','Country']].groupby(by='Country').agg({'CustomerID':'count'})
#361878
```

```python
#Which customer(CustomerID) has done highest transaction in the dataset.
df['price'] = df['UnitPrice'] * df['Quantity']
```

```python
df.columns
```

```python
df[['price','CustomerID']].groupby(by='CustomerID').agg({'price':'sum'}).sort_values('price',ascending=False)
#14646.0
```

```python
#What is the average amount of transactions:*
df['price'].mean()
#17.9
```

```python
#What is the frequently bought item in the dataset*
df_sort = df[['Description']].groupby(by='Description').agg({'Description':'count'})
df_sort.columns = ['count']
df_sort.sort_values('count',ascending=False)
#WHITE HANGING HEART T-LIGHT HOLDER
```

```python
#Create a DataFrame with unique description of item ,
#count of the item ,total amount paid for the item. Then get the least purchased item.*
#df_item = 
df_item  = df[['Description','price']].groupby(by='Description').agg({'Description':'count','price':'sum'})
df_item.columns=['count','price']
```
